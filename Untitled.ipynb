```python
#Dependencies
import pandas as pd
import requests
import json
import os
```

```python
# pulling in datasets

#cdc csv
cdc = "cdc_pre_vax.csv"

#policy data
policy = "policy_table.csv"
```

```python
cdc_df = pd.read_csv(cdc)
policy_df = pd.read_csv(policy)
```

```python
cdc_df
```

```python
policy_df
```

```python
ship_df = pd.merge(cdc_df, policy_df, on = ['Date','State'])
```

```python
ship_df
```

```python
ship_df.to_csv("shelter_in_place.csv", index=False, header=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar("State", "Total Cases", data = ship_df, color = "blue")
plt.xlabel("State")
plt.ylabel("Total Cases")
plt.title("State Comparsion")
plt.show()
```

```python
barh = ship_df.plot(kind='barh',figsize=(30, 16), color = "red", legend = None)
barh
plt.yticks(fontsize = 24)
plt.xticks(fontsize = 18)
plt.xlabel("Total Cases", fontsize = 22)
plt.ylabel("State", fontsize = 22)
plt.title("State Comparsion", fontsize=32)
barh.spines['top'].set_visible(False)
barh.spines['right'].set_visible(False)
barh.spines['bottom'].set_linewidth(0.5)
barh.spines['left'].set_visible(True)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

# Using pairplot we'll visualize the data for correlation
sns.pairplot(ship_df, x_vars=['Statw'], 
             y_vars='Total Cases', size=4, aspect=1, kind='scatter')
plt.show()
```
